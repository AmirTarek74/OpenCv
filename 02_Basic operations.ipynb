# Required Libraries

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

I will define ``imshow`` function to plot the image below the cell rather than in another window like regular opencv ``imshow`` function

```python
def imshow(img):
    '''
    function to display image using matplotlib
    Arguments:
    img : img read by opencv
    return:
    Display this image
    '''
    img_copy = img.copy()   #take copy or original image
    img_copy = cv2.cvtColor(img_copy,cv2.COLOR_BGR2RGB)    #conver from BGR (opencv defalut order) to RGB which matplot works with
    plt.imshow(img_copy)   #displayig the image
    plt.show();
```

```python
img_path = 'Nature.jpg'
img = cv2.imread(img_path) #read the image
```

```python
imshow(img) #display this image
```

kinda cool right? you can also display more than one image in the same figure using ``subplot`` but I'll leave it to you to look after it.

# Image cropping

in opencv the top left corner is our orgin point (0,0) while you are going down y is increasing and while going right x is increasing as shown in the figure below.

![Image Coordinates in opencv](coords.gif)

so to crop certain part from the image consider it slicing from a list, all you need to do is provide x range and y range.

```python
# let's first look at image shape
img_shape = img.shape
print(img_shape)
```

``shape`` attribute returns (height , width,number of channels) so we have height = 1080 pixels, width = 1920 pixels and of course 3 channels (RGB).
if we want to crop like 500 pixels from x=200 to x=700 and 500 pixels from y=300 to y=800 we will do the following.

```python
cropped_img = img[200:700,300:800]    #crop required region
```

```python
print(f"Cropped image shape :{cropped_img.shape}")
imshow(cropped_img)
```

it's really just a list indexing!

# Scaling
Scaling is just resizing of the image. OpenCV comes with a function ``cv2.resize()`` for this purpose. The size of the image can be specified manually, or you can specify the scaling factor.

```python
resized_shape = (500,500) 
resized_img = cv2.resize(img,resized_shape)    #resize orginal image to (500,500) instead of (1080, 1920)
```

```python
imshow(resized_img)
```

# Channels splitting
you can get each channel alone using ``split`` function , it will return 3 arrays __blue,green,red__ each with height and width like the original image but with only 1 channel (gray scale).

```python
blue,green,red=cv2.split(img) #splitting channels
```

```python
print(f"blue channel shape {blue.shape}")
imshow(blue)   #gray scale image
```

if you want to highlight desired channel you can use ``cv2.merge`` and put zeros in the other channels.

```python
zeros_channel = np.zeros_like(blue)  #zero array with same shape
blue_img = cv2.merge([blue,zeros_channel,zeros_channel])       #getting only blue channel color
```

```python
imshow(blue_img)
```

It's kinda scary for me. however you can do the same with red and green channels and you also can restore original image by merging all 3 channels again.
